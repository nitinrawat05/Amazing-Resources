```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simulated data
np.random.seed(0)
hours_studied = np.random.randint(1, 10, 10)  # Random hours between 1 and 9
test_scores = 5 * hours_studied + np.random.randint(-5, 5, 10)  # Linear relation with some noise

# Create a DataFrame
data = pd.DataFrame({
    'Hours Studied': hours_studied,
    'Test Score': test_scores
})

# Fit a linear regression model
model = LinearRegression()
model.fit(data[['Hours Studied']], data['Test Score'])

# Predict using the model
predicted_scores = model.predict(data[['Hours Studied']])

# Plotting the data and the regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['Hours Studied'], data['Test Score'], color='blue', label='Actual Data')
plt.plot(data['Hours Studied'], predicted_scores, color='red', label='Line of Best Fit')
plt.title('Linear Regression: Hours Studied vs. Test Score')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.legend()
plt.grid(True)
plt.show()

data
```
